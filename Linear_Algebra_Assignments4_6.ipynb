### Task 2 - Define helper functions that will help in calculating transformed vectors and also plotting them

### `We will now define two functions:`
- One to calculate a transformed vector, given an input vector and a matrix tansformation
- Another to plot the original and the transformed vectors
- In general, linear transformations can be applied to a single vector or the entire 2D plane of vectors
- In this exercise, we will visualise specific vectors and their transformed versions on the same graph
- In general, if a transformation is applied to the entire 2D plane, the grid lines would transform as well

```python
# Define a function "lineartransform()" that takes in a vector and a matrix and returns the transformed vector
# Note: Pass the input vector to the function as a list of length 2
# Note: Pass the input matrix to the function as a list of length 4
# Note: The output of the function must be a list of length 2

def lineartransform(v, t):
    # convert t into matrix and reshape it into (2,2)
    t = np.matrix(t).reshape(2, 2)

    # convert v into matrix and transpose for multiplication
    v = np.matrix(v).transpose()

    # multiply both and return w[0,0] and w[1,0]
    w = t*v
    return [w[0,0], w[1,0]]

```

```python
# Use the function "showvec()" to visualise the vector (-1, 1)
showvec([-1,1])
```

```python
# Use the function "lineartransform()" to transform the vector (-1, 1) using the matrix whose elements are [2, 0, -1, 2]
lineartransform([-1, 1], [2, 0, -1, 2])
```

```python
# Use the function "showvec()" to visualise the output vector from the "lineartransform()" function that you ran previously
# Note: You can pass the function "lineartransform()" with the required parameters as an input to the function "showvec()"
showvec(lineartransform([-1, 1], [2, 0, -1, 2]))
```

```python
# Define a function "showtransform()" that takes in a vector and a matrix and plots the original and the transformed vector
# Note: Use the "lineartransform()" function within the "showtransform()" function to get the transformed vector
# Note: Show both the original and the transformed vectors on the same figure
# Note: You may set the "color" arguments for the transformed vector as "red"
def showtransform(v, t):
    # instantiate lineartransform and pass v, t
    w = lineartransform(v, t)

    plt.figure(figsize = (6, 6))
    # plot [0, v[0]], [0, v[1]] with color='blue' and zorder=3
    plt.plot([0, v[0]], [0, v[1]], color='blue', zorder=3)

    # plot scatterplot of v[0], v[1] with color='blue' and zorder=3
    plt.scatter(v[0], v[1], color='blue', zorder=3)

    # plot [0, w[0]], [0, w[1]] with color='red' and zorder=3
    plt.plot([0, w[0]], [0, w[1]], color='red', zorder=3)

    # plot scatterplot of w[0], w[1] with color='red' and zorder=3
    plt.scatter(w[0], w[1], color='red', zorder=3)


    # keep xlim and ylim [-5,5] and xticks, yticks in range of (-5,6,1)
    plt.xlim([-5,5])
    plt.ylim([-5,5])
    plt.xticks(np.arange(-5, 6, 1))
    plt.yticks(np.arange(-5, 6, 1))
    plt.grid()
```

```python
# Use the function "showtransform()" for the vector (-2, 1) and the matrix with values [-2, 0, 1, 3]

showtransform([-2, 1], [-2, 0, 1, 3])
```

```python
# Use the function "showtransform()" for the vector (2, 0) and the matrix with values [0, 2, -1, 1]

showtransform([2, 0], [0, 2, -1, 1])
```

### **`Observations`**

- We defined a function called "lineartransform()" that:
1. Takes in an input vector as a list of length 2
2. Takes in an input matrix as a list of 4 values and reshapes it into a 2 x 2 matrix
3. Converts the input list into a vector of shape 2 x 1
4. Multiplies the vector by the matrix
5. Returns the transformed vector as a list

- We defined a function called "showtransform()" that:
1. Takes in an input vector as a list of length 2
2. Takes in an input matrix as a list of length 4
3. Uses the "lineartransform()" function to calculate the transformed vetor
4. Displays the original and the transformed vectors on the same graph

### Task 4 - Calculate eigenvalues and eigenvectors for matrix transformations

- We will now calculate eigenvalues and eigenvectors for transformation matrices using Numpy
- Although Numpy can calculate these values easily, it is recommended to brush up on the theory of eigenvalues and eigenvectors

![](https://www.sharetechnote.com/image/EngMath_Matrix_Eigen_Eq_02.PNG)

```python
# Define a transformation matrix "m" whose first row is [2, 1] and second row is [1, 3]
m = np.matrix([2, 1, 1, 3]).reshape(2, 2)
```

```python
# Print the value of "m"
print(m)
```

```python
# Use the function "np.linalg.eig()" to calculate the eigenvalues and eigenvectors of the matrix "m"
np.linalg.eig(m)
```

```python
# Store the result of "np.linalg.eig(m)" in a variable called "eig"
eig= np.linalg.eig(m)
```

```python
# Access the first element of "eig" using "eig[0]"
eig[0]

# These are the two eigenvalues
```

```python
# Access the second element of "eig" using "eig[1]"

eig[1]
```

- These are the two eigenvectors
- The eigenvectors are stored column-wise
- The first column is the first eigenvector corresponding to the first eigenvalue
- The second column is the second eigenvector corresponding to the second eigenvalue

```python
# Access the first eigenvalue using "eig[0][0]"

eig[0][0]
```

```python
# Access the seond eigenvalue using "eig[0][1]"
eig[0][1]
```

```python
# Access the first eigenvector using "[eig[1][0, 0], eig[1][1, 0]]"
[eig[1][0,0],eig[1][1,0]]
```

```python
# Access the second eigenvector using "[eig[1][0, 1], eig[1][1, 1]]"
[eig[1][0,1],eig[1][1,1]]
```

```python
# Define a function called "eigvecs()" that returns a single eigenvector of a given transformation matrix
# Note: Pass the input matrix to the function as a list of length 4
# Note: Define an additional parameter called "whichvec" to differentiate between the two eigenvectors
# Note: The output of the function must be a list of length 2

def eigvecs(m, whichvec = 2):
    # convert m into matrix and reshape (2,2)
    m = np.matrix(m).reshape(2, 2)

    # find eig value of m
    eig = np.linalg.eig(m)

    # return respective eig vector based on its which vec value
    return [eig[1][0, whichvec], eig[1][1, whichvec]]

    # setting the condition
    if whichvec == 0:
        # return respective eig vector based on its which vec value
        return [eig[1][0,0], eig[1][1,0]]

    elif whichvec == 1:
        # return respective eig vector based on its which vec value
        return [eig[1][0,1], eig[1][1,1]]

    else:
        return(print('Choose an eigenvector using "whichvec = 0" or "whichvec = 1"'))
```

```python
# Use the function "eigvecs()" to find the first eigenvector for the matrix with values [3, -1, -2, -4]
# Note: Use the "whichvec" parameter to choose your eigenvector
eigvecs([3, -1, -2, -4],0)
```

```python
# Use the function "eigvecs()" to find the second eigenvector for the matrix with values [3, -1, -2, -4]
# Note: Use the "whichvec" parameter to choose your eigenvector

eigvecs([3, -1, -2, -4],1)
```

### **`Observations`**

- We used Numpy to calculate the eigenvalues and the eigenvectors for a transformation matrix
- Since the transformation matrices in this exercise are of shape 2 x 2, we get 2 eigenvalues and 2 eigenvectors
- We understood how to extract the eigenvalues as numbers and eigenvectors as lists

### **`Observations`**

- Every matrix transformation has eigenvectors and eigenvalues
- Eigenvectors of transformation matrices are special vectors which don't change directions when the transformation is applied
- Eigenvalues are the scalar numbers by which the eigenvectors are scaled during the transformation

### **`Conclusions`**

- Eigenvectors of transformation matrices are special vectors which do not change direction during transformation
- Eigenvalues are the scalar numbers by which the eigenvectors are scaled during the transformation
- Eigenvectors can be used as reference points when performing transformations on data
